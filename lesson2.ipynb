```python
import math                       
#建立工具箱：觀望的標準函式庫有很多工具箱，看要運用什麼(如統計、數學運算等等) 如果要建立數學運算則 要輸入import math
y = float(input('請輸入對邊'))   #範例：有斜邊及對邊，要算出角度  # = 是建立變數
side = float(input('請輸入斜邊'))  
theta = math.asin( y / side)
degree = math.degrees(theta)
print("對邊:",y,"斜邊:",side,"角度:",round(degree,ndigits=2))
```

```python
import math                       
y = float(input('請輸入對邊'))   #範例：有斜邊及對邊，要算出角度  # = 是建立變數
side = float(input('請輸入斜邊'))  
theta = math.asin( y / side)
degree = math.degrees(theta)  #整合字串和數值要將字串數值格式化(format,簡稱f)->字串差補 
print(f"對邊:={y},斜邊:{side},角度:={round(degree,ndigits=2)}")
#範例 N=5 ; M=10 要轉成"N=5,M=10"  -> n=5 (換行) m=10 (換行) print(f"n={n},m={m}") -執行
```

```python
import statistics
#平均值
statistics .mean([0,5,9,14])
statistics .mean([5,6,8,9])
#算標準差 (可以知道哪個數值差距較大)
print(statistics .stdev([0,5,9,14]))
print(statistics .stdev([5,6,8,9]))
```
